## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Gradient Descent Function

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    costs = []
    ms = []
    bs = []
    
    for i in range(iterations):
        y_predict = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y - y_predict)])
        md = (-2/n) * sum(x * (y - y_predict))
        bd = (-2/n) * sum(y - y_predict)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        costs.append(cost)
        ms.append(m_curr)
        bs.append(b_curr)

        # Print every 1000 iterations
        if i % 1000 == 0:
            print(f"m = {m_curr}, b = {b_curr}, cost = {cost} iteration = {i}")
    
    return ms, bs, costs
```

## Data

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

## Performing gradient descent

```python
ms, bs, costs = gradient_descent(x, y)

# Plot the regression line
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, ms[-1] * x + bs[-1], color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()

plt.tight_layout()
plt.show()
```
